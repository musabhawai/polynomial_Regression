```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Position_Salaries.csv')
df
```

```python
x=df[["Level"]]
y=df.iloc[:,2]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
model.predict([[6]])
```

```python
y_pred=model.predict(x)
y_pred
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='blue')
plt.title('Linear regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly2=PolynomialFeatures(degree=2)
x_poly2=poly2.fit_transform(x)
x_poly2
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x_poly2,y)
model1.predict([[1,7,49]])
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,model1.predict(poly2.fit_transform(x)),color='blue')
plt.title('Polynomial Linear regression with degree 2')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly3=PolynomialFeatures(degree=3)
x_poly3=poly3.fit_transform(x)
x_poly3
```

```python
from sklearn.linear_model import LinearRegression
model2=LinearRegression()
model2.fit(x_poly3,y)
model2.predict([[1,7,49,343]])
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,model2.predict(poly3.fit_transform(x)),color='green')
plt.title('Polynomial Linear regression with degree 3')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
